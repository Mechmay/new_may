### Visualization Steps
we will be visualizing the Netflix quarterly data!

We want to get an understanding of the distribution of the Netflix quarterly stock prices for 2017. Specifically, we want to see in which quarter stock prices flucutated the most. We can accomplish this using a violin plot with four violins, one for each business quarter!

```python
plt.figure()
sns.set_style('whitegrid')
ax = sns.violinplot(data=netflix_stocks_quarterly, x= 'Quarter', y='Price', palette="husl")
ax.set_title('Distribution of 2017 Netflix Stock Prices by Quarter')
plt.xlabel('Business Quarters in 2017')
plt.ylabel('Closing Stock Price')
plt.savefig('violinplot for the Distribution of 2017 Netflix stock prices by quarter.png')
plt.show()
```

### Graph Communication
Netflix Stock Prices progressed positively throughout the year 2017 starting out with the price slightly above `$120.00` in the first quarter and closing above `$200` in the last quarter. We can also see that price ranged between `$140` to `$180` before the eventual rise in price at the end of Q4

## Step 2
we will chart the performance of the earnings per share (EPS) by graphing the estimate Yahoo projected for the Quarter compared to the actual earnings for that quarters. We will accomplish this using a scatter chart.

```python
plt.clf()
plt.figure()
x_positions = [1, 2, 3, 4]
chart_labels = ["1Q2017","2Q2017","3Q2017","4Q2017"]
earnings_actual =[.4, .15,.29,.41]
earnings_estimate = [.37,.15,.32,.41 ]
ax = plt.subplot()
plt.scatter(x_positions, earnings_actual, color='red', alpha=0.5)
plt.scatter(x_positions, earnings_estimate, color='blue', alpha=0.3)
plt.legend(['Actual', 'Estimate'], loc='right')
ax.set_xticks(x_positions)
ax.set_xticklabels(chart_labels)
plt.savefig('Scatter plot actual earnings per share vs estimate.png')
plt.show()
```

### Graph Communication
We can see from the graph above how close the estimation of EPS was and how the actual EPS did from what was expected. Did it surpass the estimated EPS by a large margin or didn’t go above the estimated EPS value? The actual EPS and estimated EPS may sometimes overlap, this data value then will be seen as a purple dot.

### Step 3
we will visualize the earnings and revenue reported by Netflix by mapping two bars side-by-side. We have visualized a similar chart in the second Matplotlib

```python
# The metrics below are in billions of dollars
plt.figure()
revenue_by_quarter = [2.79, 2.98,3.29,3.7]
earnings_by_quarter = [.0656,.12959,.18552,.29012]
quarter_labels = ["2Q2017","3Q2017","4Q2017", "1Q2018"]

ax3 = plt.subplot()
# Revenue
n = 1  # This is our first dataset (out of 2)
t = 2 # Number of dataset
d = 4 # Number of sets of bars
w = 0.8 # Width of each bar
bars1_x = [t*element + w*n for element
             in range(d)]
plt.bar(bars1_x, revenue_by_quarter)


# Earnings
n = 2  # This is our second dataset (out of 2)
t = 2 # Number of dataset
d = 4 # Number of sets of bars
w = 0.8 # Width of each bar
bars2_x = [t*element + w*n for element
             in range(d)]
plt.bar(bars2_x, earnings_by_quarter, color='red')


middle_x = [ (a + b) / 2.0 for a, b in zip(bars1_x, bars2_x)]
labels = ["Revenue", "Earnings"]

ax3.set_xticks(middle_x)
ax3.set_xticklabels(quarter_labels)
plt.legend(labels)
plt.title('Netflix Earnings and Revenue Reported by quarter')
plt.savefig('sidebyside Netflix Earnings and Revenue Reported by quarter.png')
plt.show()
```

### Grap Communication
The revenue seems to increase overtime so as the earnings and we can depict what percentage of revenue constitutes earnings for example in Q4, Netflix' earnings was `$3,700,000,000($3.7 billion)` and after deducting expenses, the net income(earnings) becomes `$290,120,000($0.29012 billion)` which is roughly 7.3% of the revenue.

### Step 4
we will compare Netflix stock to the Dow Jones Industrial Average in 2017. We will accomplish this by plotting two line charts side by side in one figure.

Since Price which is the most relevant data is in the Y axis, let's map our subplots to align vertically side by side.

```python
# Left plot Netflix
plt.figure()
ax1 = plt.subplot(2, 1, 1)
plt.plot(netflix_stocks['Date'], netflix_stocks['Price'], color='blue')
ax1.set_title('Netflix')
ax1.set_xlabel('Date')
ax1.set_ylabel('Stock Price')
ax1.set_xticks(netflix_stocks['Date'])
ax1.set_xticklabels(netflix_stocks['Date'], rotation=90)


# Right plot Dow Jones
ax2 = plt.subplot(2, 1, 2)
plt.plot(dowjones_stocks['Date'], dowjones_stocks['Price'], color='red')
ax2.set_title('Dowjones')
ax2.set_xlabel('Date')
ax2.set_ylabel('Stock Price')
ax2.set_xticks(dowjones_stocks['Date'])
ax2.set_xticklabels(dowjones_stocks['Date'], rotation=90)
plt.subplots_adjust(hspace=1.5)
plt.savefig('Netflix vs Dowjones Stockprice.png')
plt.show()

```

### Grapg Communication
The line plot represents a good way to show the trend of the price in 2017, Netflix stocks performed very well relative to Dow jones and as for volatility we can see in Q2 when Netflix stock price dropped, Dow jones kept on going up so we say the Dow jones was more volatile.
