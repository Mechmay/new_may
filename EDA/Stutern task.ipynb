```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#working dataset
df = pd.read_csv('StudentsPerformance (5) - StudentsPerformance (5).csv')
display(df.head(), df.shape)
```

```python
display(df.describe())

#Total number of participants by gender
gender = df.groupby('gender').lunch.count().reset_index()
display(gender)
```

```python
#sorting race/ethnicity by groups 
groups = df.sort_values(by='race/ethnicity')
display(groups.head())
```

```python
plt.close()
plt.figure(figsize=(12,12))
ax = plt.subplot(2,2,1)
sns.set_style('whitegrid')
sns.barplot(data = groups, x='race/ethnicity', y='math score', hue='gender', edgecolor= 'k', palette='Blues')
plt.title('Barplot of Math Scores of the different Groups')
plt.savefig('Barplot of Math Scores of the different Groups')
plt.legend(loc=8)
plt.show()

plt.close()
plt.figure(figsize=(12,12))
ax = plt.subplot(2,2,2)
sns.set_style('whitegrid')
sns.barplot(data = groups, x= 'race/ethnicity', y = 'reading score', hue = 'gender', edgecolor = 'k', palette='Greens')
plt.title('Barplot of Reading Scores of the different Groups')
plt.savefig('Barplot of Reading Scores of the different Groups')
plt.legend(loc=8)
plt.show()

plt.close()
plt.figure(figsize=(12,12))
ax = plt.subplot(2,2,2)
sns.set_style('whitegrid')
sns.barplot(data = groups, x= 'race/ethnicity', y = 'writing score', hue = 'gender', edgecolor = 'k', palette='Greys')
plt.title('Barplot of Writing Scores of the different Groups')
plt.savefig('Barplot of Writing Scores of the different Groups')
plt.legend(loc=8)
plt.show()
```

```python
#Discriptive Statistics of each group 
groupA = df[df['race/ethnicity'] == 'group A']
groupB = df[df['race/ethnicity'] == 'group B']
groupC = df[df['race/ethnicity'] == 'group C']
groupD = df[df['race/ethnicity'] == 'group D']
groupE = df[df['race/ethnicity'] == 'group E']
```

## Group A

```python
#center (mean, median) of the dataset and spread

#Math Score
math_values = groupA['math score'].values
math_mean = np.mean(math_values)
math_median = np.median(math_values)


math_quartile1 = np.percentile(math_values,25)
math_quartile3 = np.percentile(math_values,75)

math_max = np.amax(math_values)
math_min = np.amin(math_values)
math_spread = math_max - math_min
math_std = np.std(math_values)
print('The mean, median and spread of Group A on their math scores are {}, {}, {} respectively with a standard deviation of {}'.format(round(math_mean, 2), math_median, math_spread, round(math_std, 2)))
```

```python
#Reading Score
reading_values = groupA['reading score'].values
reading_mean = np.mean(reading_values)
reading_median = np.median(reading_values)

reading_max = np.amax(reading_values)
reading_min = np.amin(reading_values)
reading_spread = reading_max - reading_min
reading_std = np.std(reading_values)
print('The mean, median and spread of Group A on their reading scores are {}, {}, {} respectively with a standard deviation of {}'.format(round(reading_mean, 2), reading_median, reading_spread, round(reading_std, 2)))
```

```python
#Writing Score
writing_values = groupA['writing score'].values
writing_mean = np.mean(writing_values)
writing_median = np.median(writing_values)

writing_max = np.amax(writing_values)
writing_min = np.amin(writing_values)
writing_spread = writing_max - writing_min
writing_std = np.std(writing_values)
print('The mean, median and spread of Group A on their writing scores are {}, {}, {} respectively with a standard deviation of {}'.format(round(writing_mean, 2), writing_median, writing_spread, round(writing_std, 2)))
```

```python
#Histograms
#Histogram for math score

plt.close()
plt.figure(figsize=(7,3))
ax = plt.subplot()
plt.hist(math_values, range = (0, 100), bins = 50, edgecolor='k')
plt.axvline(math_mean, color = '#FD4E40', linewidth = 2, linestyle = 'solid', label = 'Mean')
plt.axvline(math_median, color = '#62EDBF', linewidth = 2, linestyle = 'solid', label = 'Median')
plt.axvline(math_mean + math_std, color='#FFB908', linestyle='solid', linewidth=2, label = "Standard Deviations(68%)")
plt.axvline(math_mean - math_std, color='#FFB908', linestyle='solid', linewidth=2)
plt.axvline(math_mean + math_std  * 2, color='#E9967A', linestyle='solid', linewidth=2, label = "Standard Deviations(95%)")
plt.axvline(math_mean - math_std * 2, color='#E9967A', linestyle='solid', linewidth=2)
plt.axvline(math_mean + math_std  * 3, color='#696969', linestyle='solid', linewidth=2, label = "Standard Deviations(99.7%)")
plt.axvline(math_mean - math_std * 3, color='#696969', linestyle='solid', linewidth=2)
plt.axvline(math_quartile1, color='#000000', linestyle = 'solid', linewidth=2)
plt.axvline(math_quartile3, color='#000000', linestyle = 'solid', linewidth=2)

plt.ylabel('Count')
plt.xlabel('Math Score')
plt.title('Histogram of Maths Scores by Group A')
ax.set_xticks(range(0, 101, 10))
plt.legend()
plt.show()

#Histogram of Reading score
plt.close()
plt.figure(figsize=(7,3))
ax = plt.subplot()
plt.hist(reading_values, range =(0, 100), bins = 50, edgecolor = 'k', color = 'lightskyblue')
plt.axvline(reading_mean, color = '#FD4E40', linewidth = 2, linestyle = 'solid', label = 'Mean')
plt.axvline(reading_median, color = '#62EDBF', linewidth = 2, linestyle = 'solid', label = 'Median')
plt.axvline(reading_mean + reading_std, color='#FFB908', linestyle='solid', linewidth=2, label = "Standard Deviations(68%)")
plt.axvline(reading_mean - reading_std, color='#FFB908', linestyle='solid', linewidth=2)
plt.axvline(reading_mean + reading_std  * 2, color='#E9967A', linestyle='solid', linewidth=2, label = "Standard Deviations(95%)")
plt.axvline(reading_mean - reading_std * 2, color='#E9967A', linestyle='solid', linewidth=2)
plt.axvline(reading_mean+ reading_std  * 3, color='#696969', linestyle='solid', linewidth=2, label = "Standard Deviations(99.7%)")
plt.axvline(reading_mean - reading_std * 3, color='#696969', linestyle='solid', linewidth=2)

plt.ylabel('Count')
plt.xlabel('Reading Score')
ax.set_xticks(range(0, 101, 10))
plt.title('Histograms of Reading Scores by Group A')
plt.legend()
plt.show()

#Histogram of Writing Score
plt.close()
plt.figure(figsize=(7,3))
ax = plt.subplot()
plt.hist(writing_values, range =(0, 100), bins = 50, edgecolor = 'k', color = 'grey')
plt.axvline(writing_mean, color = '#FD4E40', linewidth = 2, linestyle = 'solid', label = 'Mean')
plt.axvline(writing_median, color = '#62EDBF', linewidth = 2, linestyle = 'solid', label = 'Median')
plt.axvline(writing_mean + writing_std, color='#FFB908', linestyle='solid', linewidth=2, label = "Standard Deviations(68%)")
plt.axvline(writing_mean - writing_std, color='#FFB908', linestyle='solid', linewidth=2)
plt.axvline(writing_mean + writing_std  * 2, color='#E9967A', linestyle='solid', linewidth=2, label = "Standard Deviations(95%)")
plt.axvline(writing_mean - writing_std * 2, color='#E9967A', linestyle='solid', linewidth=2)
plt.axvline(writing_mean + writing_std  * 3, color='#696969', linestyle='solid', linewidth=2, label = "Standard Deviations(99.7%)")
plt.axvline(writing_mean - writing_std * 3, color='#696969', linestyle='solid', linewidth=2)

plt.ylabel('Count')
plt.xlabel('Writing Score')
ax.set_xticks(range(0, 101, 10))
plt.title('Histograms of Writing Scores in Group A')
plt.legend()
plt.show()
```

#### Group A Summary
Generally, All three test have a symmetrical dataset but Group A participants performed better in their math scores having its mean and median at 61.3 and 61 respectively with a Spread of 72 which is the lowest spread across the 3 tests and 95% of normally distributed values are within about 1.96 standard deviations of the mean with one outlier at 28. Group A Reading Score comes in second place with a spread of 77 with it's mean and median at 64.67 and 64 respectively with most of the data concentrated at 95% standard deviations away from the mean with 2 values at 99.7% standard deviations away from the mean at 23, 32. lastly, writing score with a spread of 78 has its mean and median at 62.67 and 62.0 respectively with most of the data concentrated at 95% standard deviations away from the mean with one value at 99.7% standard deviations away from the mean at 20.
