```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\User\Desktop\language\DATA JOURNEY\python\Data visualization_&_stats\Datasets\Life Expectancy\Life Expectancy and GDP Capstone\all_data.csv')
```

```python
display(df.head())
```

```python
display(df['Country'].unique())
df.rename(columns={'Life expectancy at birth (years)': 'Leab'}, inplace=True)
display(df.columns)
```

```python
ctry = df['Country'].unique()
def plt_country():
    for i in ctry:
        cty = df[df.Country == i]
#       ax, fig = plt.subplots(figsize=(6,6))
        plt.plot(cty.Year, cty.Leab, label = i, marker = 'o')
        plt.grid(linestyle=':')
#         plt.legend()
        plt.xlabel('Years')
        plt.ylabel('Life expectancy at birth (years)')
        plt.title('Line graph showing Life expectancy of various countries from 2000 - 2015')
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
        #bbox_to_anchor accepts a tuple of (up to) 4 values (x, y, width and height) where x and y are the lower left corner of the box
        
    plt.show()
print(plt_country())
```

Life expectancy in Zimbabwe was at a very low average over the decade in the Years 2000 to 2009 decade where Zimbabweans were on average expected to live for up to 46 years. This histrionic decline in life expectancy may be  attributed to  the deterioration of  the standard  of living  which  was  a  result  of  a decline  in  the  countrys economic activities. The Zimbabwean economy experienced a meltdown during this period. Hyperinflation coupled with poor economic management policies brought the Zimbabwean economy to near standstill. As a result of this, there were basic  food shortages and most retail commodities were found only in the  streets. The socio-economic sector was also heavily affected by the adverse economic conditions in Zimbabwe. There was a huge decline in the education  standards  and  the  health  sector  nearly collapsed. This  contributed  significantly  to  the declining  life expectancy.

However, between 2010  and 2015, life expectancy  in Zimbabwe  increased to  an average  of about 57 years. On average, Zimbabweans now lived about 11 years longer in this decade relative to the previous decade. This improvement could be credited  to the  formation of a Government of National Unity (GNU) formed between the Zimbabwe  African  National  Union-Patriotic  Front  (ZANU  PF)  and  the  two  factions  of  the  Movement  for Democratic Change (MDC). The coalition of these political parties to form a government of national unity brought stability  and  some sense  of hope  in Zimbabwe. The use  of  multiple  foreign  currencies, to  a larger  extent, also resuscitated  the  economy  which  was  on  the brink  of  total  collapse  at  the  time of  the  formation  of  the  unity government. Accordingly, state hospitals, shops and other businesses reopened and basic food staffs were back on the market. This improved the overall mortality of the Zimbabwean population. Source: <a href="https://www.researchgate.net/publication/297750732_Life_Expectancy_In_Zimbabwe_An_Analysis_Of_Five_Decades" target="_blank" rel="noopener" class="anchor__3DnlHDlaeJnh9whw0dBR8K">Life Expectancy in Zimbabwe</a>

```python
plt.close('all')
ctry = df['Country'].unique()
def plt_country():
    for i in ctry:
        cty = df[df.Country == i]
#       ax, fig = plt.subplots(figsize=(6,6))
        plt.plot(cty.Year, cty.GDP, label = i, marker = 'o')
        plt.grid(linestyle=':')
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
        plt.xlabel('Years')
        plt.ylabel('GDP')
        plt.title('Line graph showing the GDP of various countries from 2000 - 2015')
    plt.show()
print(plt_country())
```

```python
ctry = df['Country'].unique()
for i in ctry:
    cty = df[df.Country == i]
    average_leab = np.mean(cty.Leab)
    average_gdp = np.mean(cty.GDP)
    median_leab = np.median(cty.Leab)
    print('The Average life expectancy at birth (years) of ' + i + ' is {} years with its median at {} years and the Average GDP is {}.'.format(round(average_leab, 2), median_leab, average_gdp))

f, ax = plt.subplots(figsize=(10,8))
sns.set_style('whitegrid')
sns.barplot(data = df, x= 'Country', y = 'Leab', edgecolor = 'k')
plt.ylabel('Average Life expectancy at birth')
plt.title('Average Life expectancy of each country')
plt.show()

plt.close('all')
f, ax = plt.subplots(figsize=(10,8))
sns.set_style('whitegrid')
sns.barplot(data = df, x= 'Country', y = 'GDP', edgecolor = 'k')
plt.ylabel('Average GDP')
plt.title('Average GDP of each country')
plt.show()
```

```python
ctry = df['Country'].unique()
data1 = []
data2 = []
for i in ctry:
    cty = df[df.Country == i]
    data1.append(df.loc[df.Country == i]['Leab'].values)
    data2.append(df.loc[df.Country == i]['GDP'].values)
ax = plt.subplot()
plt.boxplot(data1, labels = ctry)
ax.set_xticks(range(1, 7))
ax.set_xticklabels(ctry, rotation = 90)

plt.xlabel('Country')
plt.ylabel('Life expectancy at birth(years)')
plt.title('Boxplot showing the distribution of the life expectancy of each country')
plt.show()

plt.close()
ax = plt.subplot()
plt.boxplot(data2, labels = ctry)
ax.set_xticks(range(1, 7))
ax.set_xticklabels(ctry, rotation = 90)
plt.subplots_adjust(top=0.95)
plt.xlabel('Country')
plt.ylabel('GDP')
plt.title('Boxplot showing the distribution of the GDP of each country')
plt.show()
```

The boxplot shows that Zimbabwe has the least life expectancy amongst the countries with the highest Life expectancy at 60 years and median at 47 years of age. Zimbabwe also has a poor GDP with no spread over the years. USA comes out top in GDP comparison.

#### Hypothesis Test

```python
from scipy.stats import ttest_ind
from scipy.stats import f_oneway
ctry = df['Country'].unique()
for i in ctry:
    cty = df[df.Country == i]
    tstatistic, pval = ttest_ind(cty.Leab, cty.GDP)
    if pval < 0.05:
        print('The p-value of ' + i + ' is {}, which means that the Life expectancy is dependent on the GDP'.format(pval))
```
