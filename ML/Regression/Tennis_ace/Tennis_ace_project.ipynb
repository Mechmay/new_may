### Project Goals
To create a linear regression model that predicts the outcome for a tennis player based on their playing habits. By analyzing and modeling the Association of Tennis Professionals (ATP) data, you will determine what it takes to be one of the best tennis players in the world.

### Dataset
My dataset is from the men’s professional tennis league, which is called the ATP (Association of Tennis Professionals). Data from the top 1500 ranked players in the ATP over the span of 2009 to 2017 are provided in file. The statistics recorded for each player in each year include service game (offensive) statistics, return game (defensive) statistics and outcomes.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
tennis = pd.read_csv('tennis_stats.csv')
pd.set_option('max_columns', None)
display(tennis.head())
print(tennis.shape)
```

```python
# Exploratory Data Analysis
# checking for missing values
tennis.isnull().sum()
```

```python
# identifying cells that have strong relationship
corr = tennis.corr()

display(corr)
plt.scatter(tennis['ServiceGamesPlayed'], tennis['Winnings'], alpha = 0.4)
plt.title('Service games played vs Winnings')
plt.ylabel('Winnings')
plt.xlabel('Service games played')

plt.show()
plt.clf()

plt.scatter(tennis['BreakPointsOpportunities'], tennis['Losses'], alpha = 0.4)
plt.title('Break points opportunities Vs Losses')
plt.ylabel('Losses')
plt.xlabel('Break Points Opportunities')
plt.show()
plt.clf()
```

```python
## Linear Regression
# Building a single linear regression model on the data to predict one of the outcomes
x = tennis[['ServiceGamesPlayed']]
y = tennis['Winnings']
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2, random_state=6)

# Create and train model
lr = LinearRegression()
lr.fit(x_train, y_train)
y_predict = lr.predict(x_test)

#plot predictions against actual winnings
plt.scatter(y_test, y_predict, alpha = 0.4)
plt.xlabel('Actual winnings')
plt.ylabel('Predicted winnings')
plt.title('Actual Vs predicted winnings - single feature')
plt.show()

# score model on test data
print('Predicting Wins with Service games played test score:',lr.score(x_test, y_test))
```

```python
feature = tennis[['BreakPointsOpportunities']]
outcome = tennis['Winnings']

# split train and test data
feature_train, feature_test, outcome_train, outcome_test = train_test_split(feature, outcome, train_size = 0.8, test_size = 0.2, random_state=6)

#creating and training your model
lr = LinearRegression()
lr.fit(feature_train, outcome_train)
outcome_predict = lr.predict(feature_test)

# plotting the predictions against actual winnings
plt.scatter(outcome_test, outcome_predict, alpha = 0.4)
plt.xlabel('Actual winnings')
plt.ylabel('Predicted winnings')
plt.title('Actual Vs predicted winnings - single feature')
plt.show()

# score model on test data
print('Predicting winnings based on Break points opportunities test score:',lr.score(feature_test, outcome_test))

# residual Analysis
residuals = outcome_predict - outcome_test
plt.scatter(outcome_predict, residuals, alpha = 0.4)
plt.title('Residual Analysis')
plt.show()
```

```python
feature1 = tennis[['DoubleFaults']]
outcome1 = tennis['Losses']

# Spliting the train and test data
feature_train, feature_test, outcome_train, outcome_test = train_test_split(feature1, outcome1, train_size = 0.8, test_size = 0.2, random_state=6)

#creating and trining the model
lr = LinearRegression()
lr.fit(feature_train, outcome_train)
outcome_predict = lr.predict(feature_test)

#plotting the predictions against Losses
plt.scatter(outcome_test, outcome_predict, alpha =0.4)
plt.xlabel('Actual Losses')
plt.ylabel('Predicted Losses')
plt.title('Actual Vs predicted Losses - single feature')
plt.show()

# score model on test data
print('Prediction of losses based on double faults test score:',lr.score(feature_test, outcome_test))

residuals = outcome_predict - outcome_test
plt.scatter(outcome_predict, residuals, alpha = 0.4)
plt.title('Residual Analysis')
plt.show()
```

```python
## Double Feature linear regression
# Building a multiple linear regression model on the data to predict yearly earnings
feature2 = tennis[['ServiceGamesPlayed', 'BreakPointsOpportunities']]
earning = tennis['Winnings']

# splitting the train and test data
feature2_train, feature2_test, earning_train, earning_test = train_test_split(feature2, earning, train_size = 0.8, test_size = 0.2, random_state = 6)

#creating and training the model
flr = LinearRegression()
flr.fit(feature2_train, earning_train)
earning_predict = flr.predict(feature2_test)

# plotting predictions against against actaul winnings
plt.scatter(earning_test, earning_predict, alpha = 0.4)
plt.xlabel('Actual winnings')
plt.ylabel('predicted winnings')
plt.title('Actual winnings vs predicted winnings - 2 features')
plt.show()

residual = earning_predict - earning_test
plt.scatter(earning_predict, residuals, alpha = 0.4)
plt.title('Residual Analysis')
plt.show()

# score model on test data
print('predicting winnings based on service game played and break point opportunites',flr.score(feature2_test, earning_test))
```

```python
# Creating a multiple linear regression model with a few more features
X = tennis[['FirstServe','FirstServePointsWon','FirstServeReturnPointsWon',
'SecondServePointsWon','SecondServeReturnPointsWon','Aces',
'BreakPointsConverted','BreakPointsFaced','BreakPointsOpportunities',
'BreakPointsSaved','DoubleFaults','ReturnGamesPlayed','ReturnGamesWon',
'ReturnPointsWon','ServiceGamesPlayed','ServiceGamesWon','TotalPointsWon',
'TotalServicePointsWon']]
Y = tennis[['Winnings']]

# splitting train and test data 
features_train, features_test, outcome_train, outcome_test = train_test_split(X, Y, train_size = 0.8)

# creating and training the model
mlr = LinearRegression()
mlr.fit(features_train, outcome_train)
outcome_prediction = mlr.predict(features_test)

# score model on test data
print('prediciton winnings based on multiple features test score',mlr.score(features_test, outcome_test))

plt.scatter(outcome_test, outcome_prediction, alpha = 0.4)
plt.xlabel('Actual outcome')
plt.ylabel('Predicted Outcome $y_i$')
plt.title('Actual winnings vs predicted winnings - 18 feature regression')
plt.show()
# print(f'The coefficient of Determination is: {score}')
print('The slope is:')
print(mlr.coef_)

residaual = outcome_prediction - outcome_test
plt.scatter(outcome_prediction, residuals, alpha = 0.4)
plt.show()
```
